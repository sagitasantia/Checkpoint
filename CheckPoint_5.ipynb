# **Exploratory Data Analysis**

Visualisasi diatasa merupakan tampilan dari semua bahasa dan saya akan mengambil top 10 dari atas dan top 10 dari terbawah untuk dieksekusi

```python
plt.figure(figsize=(15, 10))
top_10_languages = df['Language'].value_counts().head(10)
sns.countplot(y=top_10_languages.index, order=top_10_languages.index, palette='viridis')
plt.title('Top 10 Most Used Languages')
plt.xlabel('Count')
plt.ylabel('Language')
plt.show()
```

Visualisasi data di atas menunjukkan grafik batang yang merepresentasikan 10 bahasa yang paling umum digunakan di New York City berdasarkan jumlah penutur. Bahasa yang mendominasi adalah Lenape (Munsee), diikuti oleh Dominican Spanish, Mandarin Chinese, African-American English, Puerto Rican Spanish, Bengali, French, Sicilian, Ukrainian, dan Cantonese.

```python
plt.figure(figsize=(15, 10))
bottom_10_languages = df['Language'].value_counts().tail(10)
sns.countplot(y=bottom_10_languages.index, order=bottom_10_languages.index, palette='viridis')
plt.title('Top 10 Least Used Languages')
plt.xlabel('Count')
plt.ylabel('Language')
plt.show()
```

```python
plt.figure(figsize=(8, 8))  # Mengatur ukuran diagram lingkaran
bottom_10_languages = df['Language'].value_counts().tail(10)  # Memilih 10 bahasa yang paling jarang digunakan
plt.pie(bottom_10_languages, labels=bottom_10_languages.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', 10))  # Membuat diagram lingkaran
plt.title('Top 10 Least Used Languages')  # Menambahkan judul diagram lingkaran
plt.axis('equal')  # Memastikan diagram lingkaran terlihat berbentuk lingkaran
plt.show()
```

Grafik serta diagram pie diatas menunjukkan daftar 10 bahasa yang jarang digunakan di New York City, dilihat dari jumlah orang yang menggunakannya. Bahasa yang paling jarang digunakan adalah Judeo-Greek, kemudian diikuti oleh Judeo-Kashani, Kadazan Dusun, Juhuri, Kabardian, Kabiye, Kabli, Kabyle, Kachin, dan Ü-Tsang Tibetan.

```python
plt.figure(figsize=(12, 6))  # Mengatur ukuran diagram batang
top_10_locations = df.groupby('Location')['Language'].value_counts().sum(level=0).sort_values(ascending=False).head(10)  # Memilih 10 lokasi dengan bahasa yang paling sering digunakan
top_10_locations.plot(kind='bar', color=sns.color_palette('viridis', 10))  # Membuat diagram batang
plt.title('Top 10 Locations Using Least Common Languages')  # Menambahkan judul diagram batang
plt.xlabel('Location')  # Label sumbu x
plt.ylabel('Count')  # Label sumbu y
plt.xticks(rotation=45, ha='right')  # Mengatur rotasi dan alignment label sumbu x
plt.tight_layout()  # Mengatur layout agar tidak tumpang tindih
plt.show()
```

Visualisasi "Top 10 Least Used Languages" memberikan gambaran tentang bahasa-bahasa yang jarang digunakan dan mendekati kepunahan. Ini mencerminkan tantangan serius dalam mempertahankan keberlangsungan bahasa-bahasa ini di tengah arus globalisasi.
```python
language_counts_by_status = df.groupby('Status')['Language'].value_counts()

top_10_languages = df['Language'].value_counts().head(10)

color_scheme = {'Residential': 'blue', 'Historical': 'green', 'Community': 'yellow', 'Reviving': 'orange', 'Liturgical': 'red'}

plt.figure(figsize=(12, 8))
for status in df['Status'].unique():
    top_10_languages_status = language_counts_by_status[status].index.intersection(top_10_languages.index)
    top_10_counts_status = language_counts_by_status[status].loc[top_10_languages_status]
    if status in color_scheme:
        plt.barh(top_10_languages_status, top_10_counts_status, color=color_scheme[status], label=status)
    else:
        plt.barh(top_10_languages_status, top_10_counts_status, label=status)
plt.xlabel('Jumlah Penggunaan')
plt.ylabel('Bahasa')
plt.title('Status top 10 bahasa yang terpopuler')
plt.legend()
plt.show()
```

Visualisasi di atas menunjukkan bahwa bahasa Lenape memiliki status "reviving," menandakan bahwa bahasa ini sedang mengalami upaya pemulihan atau penyelamatan agar tetap hidup. Selanjutnya, bahasa Ukrainian memiliki status "mixed community," yang mengindikasikan bahwa penggunaannya bersifat bervariasi dan melibatkan beragam kelompok masyarakat.

Sicilian, di sisi lain, memiliki status "historical" dengan populasi pengguna sekitar 1 juta dan status "residential" dengan lebih dari 1 juta pengguna, menunjukkan sejarah serta keberadaannya yang tetap relevan dalam konteks tempat tinggal.

Bahasa French memiliki status "community" dengan jumlah pengguna sekitar 1 juta dan status "residential" dengan lebih dari 1 juta pengguna, menandakan perannya sebagai bahasa komunitas yang signifikan di New York City.

Cantonese, Puerto Rican Spanish, Mandarin Chinese, Dominican Spanish, dan African-American English semuanya memiliki status "residential" dengan lebih dari 1 juta pengguna. Ini menggambarkan bahwa bahasa-bahasa tersebut secara aktif digunakan dalam lingkungan tempat tinggal di New York City.

Bengali, seperti Sicilian, memiliki status "historical" dengan populasi pengguna sekitar 1 juta dan status "residential," menyoroti sejarah dan keberadaannya yang kuat dalam komunitas kota tersebut.

```python

language_counts_by_status = df.groupby('Status')['Language'].value_counts()

top_10_languages = df['Language'].value_counts().tail(10)

color_scheme = {'Residential': 'blue', 'Historical': 'green', 'Community': 'yellow', 'Reviving': 'orange', 'Liturgical': 'red'}


plt.figure(figsize=(12, 8))
for status in df['Status'].unique():
    top_10_languages_status = language_counts_by_status[status].index.intersection(top_10_languages.index)
    top_10_counts_status = language_counts_by_status[status].loc[top_10_languages_status]
    if status in color_scheme:
        plt.barh(top_10_languages_status, top_10_counts_status, color=color_scheme[status], label=status)
    else:
        plt.barh(top_10_languages_status, top_10_counts_status, label=status)
plt.xlabel('Jumlah Penggunaan')
plt.ylabel('Bahasa')
plt.title('Status top 10 bahasa yang hampir punah')
plt.legend()
plt.show()
```

#### **Imputasi Outliers**

Melakukan flooring mengacu pada proses mengisi nilai-nilai dalam kolom "Global Speakers" dengan nilai q1, yang merupakan kuartil pertama dari distribusi data. Tujuannya adalah untuk menangani potensi outlier dengan menetapkan batas bawah yang lebih realistis dalam analisis data. Sebaliknya, capping merujuk pada mengisi nilai-nilai dengan nilai q3, yang merupakan kuartil ketiga dari distribusi data, untuk menetapkan batas atas yang lebih realistis dalam analisis data.

```python
Q1 = df['Global Speakers'].quantile(0.25)
Q3 = df['Global Speakers'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width * IQR)
upper_whisker = Q3 + (whisker_width * IQR)
df['Global Speakers']=np.where(df['Global Speakers']>upper_whisker,upper_whisker,np.where(df['Global Speakers']<lower_whisker,lower_whisker,df['Global Speakers']))
```

###**Pre-Processing**

```python
x_final = df.drop("Unnamed: 0", axis=1)
```

menghapus kolom yang bernama "Unnamed: 0" dari data frame df. Kolom tersebut mungkin merupakan kolom yang tidak memiliki label atau nama yang jelas, atau mungkin merupakan indeks yang tidak dibutuhkan dalam analisis data lebih lanjut. #####**Normalisasi Data**

```python
scaler = MinMaxScaler()

x_final_norm = scaler.fit_transform(x_final)
```

menggunakan metode yang disebut "MinMaxScaler" untuk mengubah data yang ada dalam array x_final ke dalam rentang yang baru. Rentang ini akan berada antara 0 dan 1. Proses ini dilakukan agar data-data yang memiliki skala yang berbeda-beda bisa dibandingkan dan dianalisis dengan lebih baik.

#####**Validasi Silang**

```python
inertia_values = []

k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x_final_norm)
    inertia_values.append(kmeans.inertia_)

plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.xticks(k_range)
plt.show()
```

Pada grafik output, sumbu x menunjukkan jumlah cluster (K), sedangkan sumbu y menunjukkan nilai inertia. Inertia merupakan ukuran seberapa jauh titik data dalam suatu cluster berada dari pusat kluster tersebut. Semakin kecil nilai inertia, semakin kompak kluster tersebut. Tujuan dari analisis ini adalah untuk menemukan titik di mana penurunan nilai inertia mulai melambat secara signifikan, menunjukkan bahwa penambahan cluster tidak memberikan peningkatan yang signifikan dalam pengelompokan data. Titik ini biasanya disebut "elbow point" dan digunakan sebagai indikasi jumlah cluster optimal yang akan digunakan dalam analisis lebih lanjut.

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(x_final_norm)
```

Menggunakan algoritma KMeans untuk menganalisis data. Dengan mengatur jumlah cluster menjadi 4, kita meminta algoritma ini untuk mengelompokkan data ke dalam 4 kelompok berbeda berdasarkan kesamaan karakteristiknya. Kemudian, kita melatih algoritma ini menggunakan data yang telah dinormalisasi

#####**Hierarchical Clustering**

```python
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
```

digunakan untuk mengelompokkan data menjadi empat kelompok menggunakan metode clustering yang disebut Agglomerative Clustering. Metode ini menggunakan pengukuran jarak euclidean antara titik data untuk menentukan seberapa mirip atau berbedanya data. Kemudian, itu menggunakan pendekatan yang disebut "linkage ward" untuk menggabungkan kelompok-kelompok yang serupa secara bertahap.

## **Tuning Hypermater**

```python
param_grid = {
    'n_clusters': [3, 4, 5, 6],  # Jumlah klaster yang akan dicoba
    'init_method': ['k-means++', 'random'],  # Metode inisialisasi centroid
    'algo_type': ['lloyd', 'elkan']  # Algoritma yang digunakan
}

# Evaluasi kualitas klastering menggunakan Silhouette Score
best_score = -1
best_params = None

for n_clusters in param_grid['n_clusters']:
    for init_method in param_grid['init_method']:
        for algo_type in param_grid['algo_type']:
            kmeans = KMeans(n_clusters=n_clusters, init=init_method, algorithm=algo_type, random_state=42, n_init='auto')
            labels = kmeans.fit_predict(x_final_norm)
            silhouette_avg = silhouette_score(x_final_norm, labels)
            print("Silhouette Score:", silhouette_avg, "dengan n_clusters =", n_clusters, ", init_method =", init_method, ", algo_type =", algo_type)
            if silhouette_avg > best_score:
                best_score = silhouette_avg
                best_params = {'n_clusters': n_clusters, 'init_method': init_method, 'algo_type': algo_type}

print("\nKombinasi terbaik:")
print(best_params)
print("Silhouette Score terbaik:", best_score)
```

Kombinasi terbaik yang ditemukan dalam tuning ini menunjukkan bahwa menggunakan 6 klaster memberikan hasil klastering yang optimal untuk dataset yang digunakan, dengan metode inisialisasi centroid 'random' dan algoritma 'lloyd'. Hasil evaluasi menggunakan Silhouette Score sebesar 0.5219 menunjukkan bahwa klastering yang dihasilkan cukup baik, dengan jarak antara klaster yang terpisah secara signifikan lebih besar daripada jarak antara titik-titik dalam klaster yang sama. Ini menunjukkan bahwa data cenderung terklaster dengan baik menggunakan kombinasi parameter yang ditemukan. Oleh karena itu, kombinasi parameter ini dapat dianggap sebagai pilihan yang baik untuk melakukan klastering pada dataset yang relevan, karena hasilnya memiliki nilai Silhouette Score yang tinggi, menunjukkan bahwa klastering tersebut memberikan pemisahan yang jelas antara klaster-klaster yang dihasilkan.
