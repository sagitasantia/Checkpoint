```python
plt.figure(figsize=(12, 6))  # Mengatur ukuran diagram batang
top_10_locations = df.groupby('Location')['Language'].value_counts().sum(level=0).sort_values(ascending=False).head(10)  # Memilih 10 lokasi dengan bahasa yang paling sering digunakan
top_10_locations.plot(kind='bar', color=sns.color_palette('viridis', 10))  # Membuat diagram batang
plt.title('Top 10 Locations Using Least Common Languages')  # Menambahkan judul diagram batang
plt.xlabel('Location')  # Label sumbu x
plt.ylabel('Count')  # Label sumbu y
plt.xticks(rotation=45, ha='right')  # Mengatur rotasi dan alignment label sumbu x
plt.tight_layout()  # Mengatur layout agar tidak tumpang tindih
plt.show()
```

Visualisasi "Top 10 Least Used Languages" memberikan gambaran tentang bahasa-bahasa yang jarang digunakan dan mendekati kepunahan. Ini mencerminkan tantangan serius dalam mempertahankan keberlangsungan bahasa-bahasa ini di tengah arus globalisasi.
```python
language_counts_by_status = df.groupby('Status')['Language'].value_counts()

top_10_languages = df['Language'].value_counts().head(10)

color_scheme = {'Residential': 'blue', 'Historical': 'green', 'Community': 'yellow', 'Reviving': 'orange', 'Liturgical': 'red'}

plt.figure(figsize=(12, 8))
for status in df['Status'].unique():
    top_10_languages_status = language_counts_by_status[status].index.intersection(top_10_languages.index)
    top_10_counts_status = language_counts_by_status[status].loc[top_10_languages_status]
    if status in color_scheme:
        plt.barh(top_10_languages_status, top_10_counts_status, color=color_scheme[status], label=status)
    else:
        plt.barh(top_10_languages_status, top_10_counts_status, label=status)
plt.xlabel('Jumlah Penggunaan')
plt.ylabel('Bahasa')
plt.title('Status top 10 bahasa yang terpopuler')
plt.legend()
plt.show()
```

Visualisasi di atas menunjukkan bahwa bahasa Lenape memiliki status "reviving," menandakan bahwa bahasa ini sedang mengalami upaya pemulihan atau penyelamatan agar tetap hidup. Selanjutnya, bahasa Ukrainian memiliki status "mixed community," yang mengindikasikan bahwa penggunaannya bersifat bervariasi dan melibatkan beragam kelompok masyarakat.

Sicilian, di sisi lain, memiliki status "historical" dengan populasi pengguna sekitar 1 juta dan status "residential" dengan lebih dari 1 juta pengguna, menunjukkan sejarah serta keberadaannya yang tetap relevan dalam konteks tempat tinggal.

Bahasa French memiliki status "community" dengan jumlah pengguna sekitar 1 juta dan status "residential" dengan lebih dari 1 juta pengguna, menandakan perannya sebagai bahasa komunitas yang signifikan di New York City.

Cantonese, Puerto Rican Spanish, Mandarin Chinese, Dominican Spanish, dan African-American English semuanya memiliki status "residential" dengan lebih dari 1 juta pengguna. Ini menggambarkan bahwa bahasa-bahasa tersebut secara aktif digunakan dalam lingkungan tempat tinggal di New York City.

Bengali, seperti Sicilian, memiliki status "historical" dengan populasi pengguna sekitar 1 juta dan status "residential," menyoroti sejarah dan keberadaannya yang kuat dalam komunitas kota tersebut.

```python

language_counts_by_status = df.groupby('Status')['Language'].value_counts()

top_10_languages = df['Language'].value_counts().tail(10)

color_scheme = {'Residential': 'blue', 'Historical': 'green', 'Community': 'yellow', 'Reviving': 'orange', 'Liturgical': 'red'}


plt.figure(figsize=(12, 8))
for status in df['Status'].unique():
    top_10_languages_status = language_counts_by_status[status].index.intersection(top_10_languages.index)
    top_10_counts_status = language_counts_by_status[status].loc[top_10_languages_status]
    if status in color_scheme:
        plt.barh(top_10_languages_status, top_10_counts_status, color=color_scheme[status], label=status)
    else:
        plt.barh(top_10_languages_status, top_10_counts_status, label=status)
plt.xlabel('Jumlah Penggunaan')
plt.ylabel('Bahasa')
plt.title('Status top 10 bahasa yang hampir punah')
plt.legend()
plt.show()
```

Dalam visualisasi data di atas, terlihat status 10 bahasa yang hampir punah, antara lain Kachin, Kabli, Kadazan Dusun, Judeo-Greek, U-Tsang Tibetan, Kabyle, Kabiye, Kabardian, Juhuri, dan Judeo-Kashani. Bahasa Kachin dan Kabli ditunjukkan dengan warna kuning, menandakan status Community, yang berarti bahasa tersebut digunakan dalam lingkungan kecil atau komunitas tertentu, dengan jumlah pengguna kurang dari 1 juta. Selanjutnya, Kadazan Dusun dan Judeo-Greek ditunjukkan dengan warna hijau, menandakan status Historical, yang berarti bahasa tersebut digunakan secara terbatas dalam konteks sejarah, juga dengan jumlah pengguna kurang dari 1 juta. Dengan demikian, visualisasi ini memberikan gambaran tentang bahasa-bahasa dengan status dan penggunaan yang semakin langka di dunia saat ini.

```python
language_counts_by_size = df.groupby(['Size', 'Language']).size().unstack().fillna(0)
top_10_languages = df['Language'].value_counts().head(10)

plt.figure(figsize=(12, 8))
for size in df['Size'].unique():
    top_10_languages_size = language_counts_by_size.loc[size].index.intersection(top_10_languages.index)
    top_10_counts_size = language_counts_by_size.loc[size].loc[top_10_languages_size]
    plt.barh(top_10_languages_size, top_10_counts_size, label=size)
plt.xlabel('Jumlah Penggunaan')
plt.ylabel('Bahasa')
plt.title('Setiap Ukuran (Size) Top 10 Bahasa yang sering Digunakan ')
plt.legend()
plt.show()
```

# **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'object'])

for col in cols:
    if df[col].dtype == 'object':
        # Jika tipe data kolom adalah objek, lewati
        continue

    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    percent_outliers = (len(outliers) / len(df)) * 100
    results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Output di atas memberikan presentase outliers dari Global Speakers, yang diperkirakan sekitar 13.385826%. Outliers adalah data yang jauh dari nilai-nilai lain dalam kumpulan data, dan mereka dapat memberikan informasi berharga tentang variasi dan distribusi data. Dalam konteks ini, Global Speakers merupakan kategori atau variabel yang memiliki data yang signifikan dan berbeda secara signifikan dari yang lain dalam kumpulan data. Karena tipe data dari kolom lainnya adalah objek, kemungkinan besar tidak ada informasi yang relevan untuk mengevaluasi outliers dalam konteks tersebut.

Mengapa hanya Global Speakers yang muncul dalam analisis outliers? Karena variabel Global Speakers mungkin merupakan variabel numerik yang memiliki distribusi data yang jelas, sehingga mungkin untuk mengidentifikasi outliers berdasarkan statistik deskriptif atau teknik lainnya. Di sisi lain, kolom lain dengan tipe data objek mungkin berisi kategori, teks, atau informasi lain yang sulit untuk dianalisis secara numerik. Oleh karena itu, sulit untuk mengidentifikasi outliers dalam konteks tersebut.
